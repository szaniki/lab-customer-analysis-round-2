```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('marketing_customer_analysis.csv') 
pd.set_option('display.max_columns', None)  # # just to see all columns in one

display(dataset)
```

```python
# standardize header names

cols = []  
for column in dataset.columns:
    cols.append(column.lower())
dataset.columns = cols
dataset
```

```python
# Which columns are numerical?
dataset.dtypes

dataset.select_dtypes(np.number).columns
```

```python
# Which columns are categorical

dataset.select_dtypes(object).columns
```

```python
# Check the occurancy of NaN values / column

dataset.isna().sum()
```

```python
# removing vehicle tpye column as half of it is missing, makes no sense to work with that much missing values

dataset = dataset.drop(['vehicle type'], axis=1)
dataset
```
